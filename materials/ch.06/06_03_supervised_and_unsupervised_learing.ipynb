### Supervised Learning

**Supervised Learning** involves training a machine learning model on a ***labeled dataset***, where each training example includes an input vector and the corresponding output label. The model learns to map inputs to the correct outputs, aiming to predict the labels for new, unseen data accurately. It is commonly used for tasks such as regression, where the output is a continuous value, and classification, where the output is a label from a predefined set of categories.

### Unsupervised Learning

**Unsupervised Learning**, in contrast to supervised learning, involves training a model using information that is neither classified nor labeled. This approach allows the model to act on the data without guidance, finding patterns and relationships on its own. It is used for exploratory data analysis, clustering, and reducing the dimensionality of data. Common applications include customer segmentation and anomaly detection.


### Unsupervised Learning Example: K-Means Clustering

In this example, we will use the K-means algorithm to cluster a dataset. K-means is a popular clustering algorithm that partitions the dataset into K distinct, non-overlapping clusters. It tries to make the intra-cluster points as similar as possible while keeping the clusters as different as possible. It assigns points to clusters based on the distance from centroids, the center of each cluster.


```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Apply K-means clustering
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plotting the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)
plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
